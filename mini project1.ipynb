```python
AutoType=[]
for row in mtcar.am :
    if row== 0:
        AutoType.append('automatic')
    elif row == 1:
        AutoType.append('manual')
mtcar['AutoType']= AutoType
mtcar.head()
```

```python
mtcar.mpg.describe()
```

```python
MileageType=[]
for row in mtcar.mpg :
    if row>23:
        MileageType.append('high mileage')
    elif row > 15:
        MileageType.append('medium mileage')
    else:
        MileageType.append('low mileage')
mtcar['MileageType']= MileageType
mtcar.head(10)
```

```python
mtcar.wt.describe()
```

```python
WtType=[]
for row in mtcar.wt :
    if row>3.61:
        WtType.append('heavy wt car')
    elif row > 2.58:
        WtType.append('medium wt car')
    else:
        WtType.append('low wt car')
mtcar['WtType']= WtType
mtcar.tail(10)
```

```python
mtcar.mpg.groupby(mtcar.AutoType).mean()
```

```python
auto= mtcar[mtcar.AutoType=='automatic']
manual = mtcar[mtcar.AutoType=='manual']
```

```python
from scipy import stats as st
```

```python
st.ttest_ind(auto.mpg,manual.mpg)
```

```python
thirdgear= mtcar[mtcar.gear==3]
fourthgear = mtcar[mtcar.gear==4]
fifthgear = mtcar[mtcar.gear==5 ]

```
